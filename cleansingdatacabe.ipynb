##### PREPARE LOOKUP TABLE FOR KABUPATEN/KOTA AND PROVINSI

```python
# PREPARE LOOKUP TABLE FOR KABUPATEN/KOTA AND PROVINSI

#Get lookup table for kota/kabupaten and provinsi
pwd = os.getcwd()
df_regions = pd.read_csv(pwd + '/dataset/regions.csv')

# Keep only two columns
df_regions = df_regions[['kab_kota', 'province']]

# Remove "Kab. " and "Kota" from the kab_kota column
df_regions['kab_kota'] = df_regions['kab_kota'].str.replace('Kab. ', '').str.replace('Kota ', '')

# Remove "Prov. " from the province column
df_regions['province'] = df_regions['province'].str.replace('Prov. ', '')

# Rename the column
df_regions = df_regions.rename(columns={'kab_kota': 'Kabupaten/Kota', 'province': 'Provinsi'})

df_regions
```

```python
# PREPARE ADDITIONAL LOOKUP TABLE FOR KABUPATEN/KOTA AND PROVINSI

#Get lookup table for kota/kabupaten and provinsi tambahan
pwd = os.getcwd()
df_regions_tambahan = pd.read_excel(pwd + '/dataset/regions_tambahan.xlsx')
df_regions_tambahan
```

```python
# COMBINE LOOKUP TABLE

# Concatenate the DataFrames vertically
combined_df_regions = pd.concat([df_regions, df_regions_tambahan])

# Reset the index of the combined DataFrame
combined_df_regions = combined_df_regions.reset_index(drop=True)

# Remove non-unique values
combined_df_regions = combined_df_regions.drop_duplicates(subset='Kabupaten/Kota')

combined_df_regions
```

##### MERGE DATASET MERGED WITH LOOKUP TABLE TO ADD PROVINSI COLUMN

```python
# Merge the DataFrames to add the new column Provinsi
merged_df = merged_df.merge(combined_df_regions, on='Kabupaten/Kota', how='left')

# Replace '-' values with NaN 
merged_df['Sambal jadi'] = merged_df['Sambal jadi'].replace('-', np.nan)
merged_df['Bumbu maskan jadi/kemasan, bumbu racikan'] = merged_df['Bumbu maskan jadi/kemasan, bumbu racikan'].replace('-', np.nan)
merged_df['Cabe merah'] = merged_df['Cabe merah'].replace('-', np.nan)
merged_df['Cabe hijau'] = merged_df['Cabe hijau'].replace('-', np.nan)
merged_df['Cabe rawit'] = merged_df['Cabe rawit'].replace('-', np.nan)

merged_df
```

```python
# CEK FOR NAN VALUE IN COLUMN PROVINSI

# Filter the DataFrame for rows with NaN values in the "Provinsi" column
df_nan_provinsi = merged_df[merged_df['Provinsi'].isna()]

# Print the 'Kabupaten/Kota'
print(df_nan_provinsi['Kabupaten/Kota'])
```
